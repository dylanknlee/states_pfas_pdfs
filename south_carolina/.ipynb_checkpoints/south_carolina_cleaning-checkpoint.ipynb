```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("south_carolina1.csv")

# Get the name of the last column
last_column = df.columns[-1]
first_column = df.columns[0]

# Drop the last column
df = df.drop(last_column, axis=1)
df = df.drop(first_column, axis = 1)

analytes = df.columns[4:]
non_analytes = df.columns[:4]

df = pd.melt(df, id_vars=non_analytes, value_vars=analytes, 
                    var_name='analyte', value_name='concentration')
df
```
