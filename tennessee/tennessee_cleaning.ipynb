```python
df1 = pd.read_csv("tennessee1.csv")
analytes = df1.columns[7:]
non_analytes = df1.columns[:7]

df1 = pd.melt(df1, id_vars=non_analytes, value_vars=analytes, 
                    var_name='analyte', value_name='concentration')
df1 = df1.sort_values(by = "Unnamed: 0").reset_index(drop = True)
df1.head(10)
```
