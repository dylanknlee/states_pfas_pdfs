```python
df3 = pd.read_csv("new_mexico3.csv", dtype={"USGS Site ID": str})
df3.drop(columns = 'Unnamed: 0', inplace = True)

# Drop the last two columns
last_two_columns = df3.columns[-2:]
df3.drop(last_two_columns, axis=1, inplace = True)

non_analytes = df3.columns[:4]
analytes = df3.columns[4:]

def id_fix(idd):
    if len(idd) < 15:
        idd = "0"+idd
    return idd.split('.0')[0]
    
df3["USGS Site ID"] = df3["USGS Site ID"].apply(id_fix)

df3 = pd.melt(df3, id_vars=non_analytes, value_vars=analytes, 
                    var_name='analyte', value_name='concentration')

df3
```

- does usgs site id = pwsid? - not the same, but can be used to find the pwsid
- can we find a more general location? some are general, others are too specific to determine coordinates
    - could use county
